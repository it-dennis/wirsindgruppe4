```python
#Thema: Listen

c64_games = ["giana sisters", "frogger", "moon patrol"]

print(c64_games)
print(c64_games[0])

# methoden

print(len(c64_games)) #len (engl) von length, deutsch = Länge
print(dir(c64_games))
print(help(c64_games))


#Listeneinträge hinzufügen

c64_games.append("terramex")

#Listeneintrag löschen
c64_games.append("Fehler")
c64_games.pop() #löscht default den letzten Eintrag, es kann aber auch der index angegeben werden.

c64_games.append("Eingabe")
c64_games.insert(2, "lotus Esprit Turbo Challenge") #Einfügen an index X, alle danach werden um 1 Stelle verschoben.

c64_games.remove("Eingabe") # löscht den ersten passenden Eintrag


print(c64_games)

c64_games.sort() # Liste alphabetisch sortieren

print(c64_games)

c64_games.reverse() #umkehren der Reihenfolge

print(c64_games)

print(sorted(c64_games)) #bindet die Sortierung in print ein
print(sorted(c64_games, reverse=True)) #erzeugt das gleiche wie c64_games.reverse(), kehrt also die Reihenfolge um

c64_games.clear() #löscht den Inhalt der Liste. Die Liste selbst bleibt bestehen, wird leer ausgeworfen.

print(c64_games)

c64_games = ["giana sisters", "frogger", "moon patrol", "terramex", "lotus Esprit Turbo Challenge"]
print(sorted(c64_games))





```

```python
#Thema: Tuple (nicht veränderbare Liste)

#Reihenfolge bleibt, schneller und speichereffizienter als Listen, Werte können nicht geändert werden (immutable)

#Gut für feste Daten, z.B. Koordinaten, Einstellungen, Rückgabewerte

farben = ("rot", "gruen", "blau") #ACHTUNG: Die Klammern unterscheiden Liste und Tuple, Liste -> liste[], tuple()

print(farben[1])


#Iteration
for f in farben:
    print(f)

#methoden funktionieren gleich bzw. ähnlich
print(len(farben))
print(dir(farben))
print(help(farben))

#sprich: Alles was mit aufrufen, anschauen usw. zu tun geht bei tuple genau gleich wie bei liste. Alles was eine Änderung bedingt, also hinzufügen, löschen
#usw. geht bei tuple nicht.


```

```python
#Thema: set

c64_games = {"giana sisters", "frogger", "frogger", "moon patrol"}  # geschwungene Klammern

print(c64_games)

#Eigenschaften Sets: Veränderung möglich, sehr schnelle Prüfung (in), keine Duplikate, keine feste Reihenfolge / kein Indexzugriff

#ideal zum entfernen von Duplikaten und schnellen "ist drin?"-Checks

#Beispiel, es werden die Variablen a und b erzeugt, jeweils mit set-listen und der Angabe von Zahlen. Die Zahl 3 ist bei beiden drin.
a = {1, 2, 3}
b = {3, 4}

#über den Befehl. a.union(b) werden a und b (also die beiden Listen) zusammengefügt.  Hierbei wird nun automatisch 3 als Duplette erkannt und 
#nur noch 1x beigefügt.
print(a.union(b))




```
